### Data Visualization

```python
import matplotlib.pyplot as plt
```

### It display the pie chart for version and score

```python
df=pd.DataFrame(data.head(500))    # Asiigning data to df variable

df['math score'].value_counts().plot(kind='pie',autopct='%1.1f%%') # creating a pie chart for score
plt.title("reading score")
#legend properties
plt.legend(loc='center', fontsize=16, shadow=True,bbox_to_anchor=(2,1))
plt.show()
```

### It displays the Bar Chart for score

```python
df=pd.DataFrame(data.head(800))
df['math score'].value_counts().plot(kind='bar',color='purple',)
plt.xlabel('reading score')
plt.ylabel('writing score')
plt.title("title")
plt.xticks(rotation=60)
plt.show()
```

### Histogram for SCORE

```python
df = pd.DataFrame(data.head(800))  # Create a DataFrame from the first 500 rows of 'data'

# Choose the column you want to visualize (replace 'column_name' with the actual name)
column_to_plot = 'math score'

# Create the histogram
plt.hist(df['math score'])
plt.xlabel('writing score')  # Label for the x-axis
plt.ylabel('')  # Label for the y-axis
plt.title('Histogram graph - X- ' + column_to_plot)  # Title for the plot
plt.grid(axis='y')  # Add gridlines for better readability (optional)
plt.show()
```

### 2D Line Graph

```python
df = pd.DataFrame(data.head(1000))  # Create a DataFrame from the first 500 rows of 'data'

# Choose the columns you want to plot 
column_to_plot_x = 'math score'
column_to_plot_y = 'reading score'

# Create the line graph
plt.plot(df['math score'], df['reading score'])
plt.xlabel('reading score')  # Label for the x-axis
plt.ylabel('math score')  # Label for the y-axis
plt.title('2D Line Graph')  # Title for the plot
plt.grid(True)  # Add gridlines for better readability
plt.show()
```

### Scatter Graph for Version & ID

```python
df = pd.DataFrame(data.head(500))  # Create a DataFrame from the first 500 rows of 'data'

# Choose the columns you want to plot (replace 'column1' and 'column2' with actual names)
column_to_plot_x = 'math score'
column_to_plot_y = 'writing score'

# Create the scatter graph
plt.scatter(df['writing score'], df['math score'])
plt.xlabel('math score')  # Label for the x-axis
plt.ylabel('writing score')  # Label for the y-axis
plt.title('Scatter Graph')  # Title for the plot
plt.grid(True)  # Add gridlines for better readability
plt.show()
```
